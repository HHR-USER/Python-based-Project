Function  
Factorial of a Number
Factorial of a number is the product of all the positive integers less than or equal to the given number

```python
number = int(input(' Enter a number to find its factorial : '))
factorial = 1
if number == 0:
  print(' The Factorial of 0 is 1')
else:
  for i in range(1, number+1):
    factorial = factorial*i
  print('The factorial of ',number,' is',factorial)
```

Factoraial Function

```python
def factorial_value(num):
  factorial = 1
  if num == 0:
    return factorial
  else:
    for i in range(1, num+1):
      factorial = factorial*i
    return factorial
print(factorial_value(5))
```

```python
# importing matplotlib library
import matplotlib.pyplot as plt
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
```

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
